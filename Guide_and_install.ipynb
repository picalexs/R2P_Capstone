### Challenge 1: QMOO (100 points)

**The maximum grade for this exercise is 100 points**

*Summary:*

In this challenge, we will tackle a multi-objective optimization (MOO) problem, reformulated as a weighted Max-Cut problem that can be solved by the Quantum Approximate Optimization Algorithm (QAOA). In particular your task is to to prepare a suitable ansatz for solving the multi-objective Max-cut problem, first for a problem with 42 qubits and then you will be asked to scale it to 80 qubits.

### Challenge 2: Implicit Solvent (100 points)

**The maximum grade for this exercise is 100 points**

*Summary:*

In this challenge, we will estimate the ground state of a Methylamine, $CH_{3}NH_{2}$  molecule using a Sample-based Quantum Diagonalization with 30 qubits that outperforms Hartree-Fock and CASCI classical solvers.

### Challenge 3: Hadron-Schwinger (90 points)

**The maximum grade for this exercise is 90 points**

*Summary:*

In this challenge, we will simulate the propagation of a hadron wavepacket in the Schwinger model on an IBM quantum computer, prepared on a lattice with $L=34$ spatial sites (68 qubits). We will start with initial states already prepared by SC-ADAPT-VQE and perform the time evolution of the Hamiltonian using the Suzuki-Trotter algorithm to evolve the initial states forward in time. The circuits will first be executed on a simulator and later on an IBM quantum computer.
